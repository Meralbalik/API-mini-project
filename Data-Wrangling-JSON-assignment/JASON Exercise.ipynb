Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

## JSON exercise work & answers

```python
import pandas as pd
```

```python
import json
from pandas.io.json import json_normalize
```

```python
# load json file as string
json_str = json.load((open( 'world_bank_projects.json')))
```

```python
# load json file as Pandas dataframe
json_df = pd.read_json('world_bank_projects.json')
json_df.head()
```

```python
#list the column names of the dataframe
json_df.columns
```

## Question 1 work:

```python
# extract the country name and project name columns as a Dataframe
country_projects = json_df[['countryname','project_name']]

# look at the first ten rows of the dataframe
country_projects.head(10)
```

```python
json_df.info()
```

## Answer 1:

## 10 countries with most projects

```python
json_df['countryname'].value_counts().head(10)
```

## Question 2 work:

```python
# extract the 'mjtheme_namecode' column 
json_df[['mjtheme_namecode']].head(20)
```

```python
#use json_normalize function to access nested information

json_norm = json_normalize(json_str, 'mjtheme_namecode')
```

```python
json_norm.info()
```

```python
json_norm.head(20)
```

## Answer 2:

## 10 major project themes

```python
json_norm['name'].value_counts().head(10)
```

## Question 3 work:

```python
# create a dict consist of names for each codes
new_json_norm={}

for code, name in zip(json_norm['code'],json_norm['name']):
    if name != '' :
        new_json_norm[code] = name
        
new_json_norm
```

## Answer 3:
##  Dataframe with the missing names filled in

```python
#fill missing names in json_norm using the names collected in new_json_norm
json_norm['name']=json_norm['code'].map(new_json_norm)
json_norm.head(10)
```

```python
# reprint the top ten project themes
json_norm['name'].value_counts().head(10)
```
